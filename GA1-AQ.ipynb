# Introduction

Last year, air quality received a lot of media attention. In May, the Dutch Coucil of State (Raad van State) ruled that a compensation program for Nitrogen production was in conflict with European law. This ruling impacted many industries, including the construction and agriculture industry.

With the economy and population growing, one may expect air quality to get worse over time. There are, however, also positive developments. In particular, electric cars are becoming more and more popular. Suppose at some point all cars would be electric, would that have a positive impact on air quality?

In this assignment, you will investigate whether there is a relationship between air quality and traffic, based on data measuring air quality and traffic in the Netherlands in April 2017. We chose the month of April 2017 since there were not many electric cars yet back then, there is no influence of daylight savings time, and the weather can be expected to be moderate as it is in spring and not in the summer or winter.

The air quality data was retrieved from [**luchtmeetnet.nl**](https://luchtmeetnet.nl). This website publishes the concentrations of various chemical compounds from approximately one hundred measurement stations in the Netherlands.

The traffic data is an aggregation of data downloaded from the [**ndw**](https://www.ndw.nu/en/), the national data warehouse on traffic information. This website stores historical traffic data measured at thousands of different locations in the Netherlands.

# Description of the dataset

The dataset consists of five files: `aq.csv`, `traffic.csv`, `closest_traffic_stations.csv`, `aq_station_info.csv` and `traffic_station_info.csv`. They are stored in the `datasets` folder. The first two files contain air-quality data and traffic data respectively. The file `closest_traffic_stations.csv` stores which traffic stations are closest to the air quality stations, and the files `aq_station_info.csv` and `traffic_station_info.csv` contain information about the measurement stations.

###  The `aq.csv` file

This file contains air quality measurements from various stations in the Netherlands. Every row in the file contains measurements of chemical compounds averaged over a certain hour in time. The start of that hour is recorded in the `csv` file. The `csv` file contains the following columns:

|Column  | Type | Description | 
|:-------|:----|:----|
|id | int | the id of the air-quality measurement station |
|local_time | string | the starting time of the hour-long averaging period measurement in local (Amsterdam) time |
|hour | int | the hour on the day in local time |
|weekday | int | the weekday encoded as an integer, with Monday = 0 |
|BC | float | black carbon concentration in $\mu \mathrm{g / m}^3$ |
|C6H6 | float | benzene ($\mathrm{C}_6 \mathrm{H}_6$) concentration in $\mu \mathrm{g / m}^3$ |
|C7H8   | float | toluene ($\mathrm{C}_7 \mathrm{H}_8$) concentration in $\mu \mathrm{g / m}^3$ |
|C8H10    | float | xylene ($\mathrm{C}_8 \mathrm{H}_{10}$) concentration in $\mu \mathrm{g / m}^3$ |
|CO | float | carbon monoxide ($\mathrm{CO}$) concentration in $\mu \mathrm{g / m}^3$ |
|H2S | float | hydrogen sulfide ($\mathrm{H}_2 \mathrm{S}$) concentration in $\mu \mathrm{g / m}^3$|
|NH3 | float | ammonia ($\mathrm{N}\mathrm{H}_3$) concentration in $\mu \mathrm{g / m}^3$ |
|NO | float | nitrogen monoxide ($\mathrm{N} \mathrm{O}$) concentration in $\mu \mathrm{g / m}^3$ |
|NO2 | float | nitrogen dioxide ($\mathrm{N} \mathrm{O}_2$) concentration in $\mu \mathrm{g / m}^3$ |
|NOx | float | nitrogen oxides concentration in $\mu \mathrm{g / m}^3$ |
|O3 | float | ozone ($\mathrm{O}_3$) concentration in $\mu \mathrm{g / m}^3$ |
|PM10 | float | particle matter (PM10) concentration in $\mu \mathrm{g / m}^3$ |
|PM25 | float | particle matter (PM2.5) concentration in $\mu \mathrm{g / m}^3$ |
|SO2 | float | sulfur dioxide ($\mathrm{S} \mathrm{O}_2$) concentration in $\mu \mathrm{g / m}^3$ |
|SOX | float | sulfur oxides concentration in $\mu \mathrm{g / m}^3$ |


Different stations may measure different chemical compounds, resulting in rows with a `NaN` for some of the compounds.

###  The `traffic.csv` file

This file contains traffic measurements from various stations in the Netherlands. Every row contains data about a single measurement. The `csv` file contains the following columns:

|Column  | Type | Description | 
|:-------|:----|:----|
| id | string | The id of the traffic measurement station |
| local_time | string | The starting time of the hour under consideration in local (Amsterdam) time |
|hour | int | the hour on the day in local time |
|weekday | int | the weekday encoded as an integer, with Monday = 0 |
| nr_of_vehicles | int | The number of vehicles passing by in the given hour|
| speed | float | The average speed of the cars in the given hour in km per hour |

### The `closest_traffic_stations.csv` file

Air quality and traffic are measured at different stations. The `closest_traffic_stations.csv` file matches air quality stations with their closest traffic stations. Every row matches an air quality measurement station with five closest traffic measurement stations:

|Column | Type | Description | 
|:-------|:----|:----|
| aq_id | int | The id of the air quality measurement station |
| traffic_id_1 | string | The id of the closest traffic measurement station |
| traffic_id_2 | string | The id of the second closest traffic measurement station |
...
| traffic_id_5 | string | The id of the tenth closest traffic measurement station |

### The `aq_station_info.csv` file

The `aq_station_info.csv` file contains the name, latitude and longitude of the air-quality measurement stations.

|Column  | Type | Description | 
|:-------|:----|:----|
| id | int | The id of the air-quality measurement station |
| name | string | The name of the air-quality measurement station |
| latitude | float | The latitude of the air-quality measurement station |
| longitude | float | The longitude of the air-quality measurement station |

### The `traffic_station_info.csv` file

The `traffic_station_info.csv` file contains the name, latitude and longitude of the traffic measurement stations.

|Column  | Type | Description | 
|:-------|:----|:----|
| id | string | The id of the traffic measurement station |
| name | string | The name of the traffic measurement station |
| latitude | float | The latitude of the traffic measurement station |
| longitude | float | The longitude of the traffic measurement station |
